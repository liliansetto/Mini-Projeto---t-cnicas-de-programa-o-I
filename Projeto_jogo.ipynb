```python
import numpy as np

# Desafio 1: Simulação de Dados
# Lançamento de dois dados de seis lados (valores de 1 a 6).
def simulacao_dados():
    dado1 = np.random.randint(1, 7) # A função random atribui o numero aleatório
    dado2 = np.random.randint(1, 7)
    print(f"Números dos dados: {dado1}, {dado2}")
    return dado1 + dado2

# Desafio 2: Múltiplas Simulações
# Número de simulações desejadas
num_simualacoes = 1000

# Lista para armazenar os resultados
resultados = [] #lista vazia para garantir que não haverá números

# Loop for para simular os lançamentos de dados
for jogada in range(num_simualacoes):
    resultado_jogada = simulacao_dados() # quantidade de vezes que os dados foram jogados

    # Armazena os resultados dos jogos num array
    resultados.append(resultado_jogada) #guarda os resultados dos jogos na variável resultados

print("\n")

print(resultados)

print("\n")

# Desafio 3: Análise de Dados
print(f"Média dos resultados: {np.mean(resultados)}") # a funçao mean calcula a média
print(f"Lançamento máximo: {np.max(resultados)}")
print(f"Lançamento mínimo: {np.min(resultados)}")

print("\n")

# Desafio 3: Número de vezes que cada lançamento ocorreu
# Criação do dicionario Ocorrências vazio
ocorrencias = {}

# Percorre a lista de 2 a 12, excluindo 0 e 1
for valor in range(2, 13):
    # Garante que todas as contagens estão zeradas
    ocorrencias[valor] = 0

for resultado in resultados:#começa a contagem da qtde de resultados
    # Conta quantas vezes cada resultado específico ocorreu nos lançamentos de dados.
    ocorrencias[resultado] += 1

# Imprimi o número de vezes que cada lançamento ocorreu
for numero in range(2, 13):
    print(f"Número de vezes que {numero} ocorreu: {ocorrencias[numero]}")


# Teste de Hipótese sem o uso da classe Counter
def teste_hipotese(resultados):
    probabilidade_esperada = 1/36  #  Cada combinação possível de dois dados (de 1 a 6 em cada dado) tem a mesma probabilidade de ocorrer, ou seja, 1/36.

    # Cria um dicionario vazio
    contagem_resultados = {}

    # Preenche o dicionário contagem_resultados com chaves de 2 a 12 (resultados possíveis de lançamentos de dois dados) e inicializa cada valor como 0.
    for i in range(2, 13):
        contagem_resultados[i] = 0

    # Conta as ocorrências de cada resultado
    for resultado in resultados:
        contagem_resultados[resultado] += 1

    print("\n Teste de Hipótese:")

    # Calcula e imprime a frequência relativa de cada resultado
    for i in range(2, 13):
        # A frequência relativa é obtida dividindo o número de ocorrências de um resultado pelo total de jogos simulados.
        frequencia = contagem_resultados[i] / len(resultados)
        print(f"{i}: Resultado: {frequencia:.4f} (Esperado: {probabilidade_esperada:.4f})")

# Teste de Hipótese
teste_hipotese(resultados)
```
